```python
import networkx as nx

G=nx.Graph()

filename = "assignp1500.txt"
f = open("assignp1500.txt","r")
n = int(f.readline())
print n
for i in f.readlines() :
    s = i.split(' ')
    a,b,w = int(s[1]),int(s[2]),int(s[3])
    G.add_edge(a,b)
    G[a][b]['weight'] = w
print G.number_of_nodes()


```
